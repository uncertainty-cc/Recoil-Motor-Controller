# FOC Math

```python
import numpy as np
from matplotlib import pyplot as plt

from matplotlib import animation
from IPython.display import HTML

plt.rcParams['figure.figsize'] = [16, 10]
```

```python
def FOC_clarkTransform(i_a, i_b, i_c):
    i_alpha  = (1. / 3.)           * (2. * i_a - i_b - i_c)
    i_beta   = (1. / np.sqrt(3.))  * (i_b - i_c)
    return i_alpha, i_beta

def FOC_parkTransform(i_alpha, i_beta, sin_theta, cos_theta):
    i_q  = -(sin_theta * i_alpha) + (cos_theta * i_beta)
    i_d  =  (cos_theta * i_alpha) + (sin_theta * i_beta)
    return i_q, i_d

def FOC_invParkTransform(v_q, v_d, sin_theta, cos_theta):
    v_alpha  = -(sin_theta * v_q) + (cos_theta * v_d)
    v_beta   =  (cos_theta * v_q) + (sin_theta * v_d)
    return v_alpha, v_beta

def FOC_invClarkSVPWM(v_alpha, v_beta):
    v_a_phase = v_alpha
    v_b_phase = (-.5 * v_alpha) + ((np.sqrt(3.)/2.) * v_beta)
    v_c_phase = (-.5 * v_alpha) - ((np.sqrt(3.)/2.) * v_beta)

    v_neutral = .5 * (np.maximum(np.maximum(v_a_phase, v_b_phase), v_c_phase)
                      + np.minimum(np.minimum(v_a_phase, v_b_phase), v_c_phase))

    v_a = v_a_phase - v_neutral
    v_b = v_b_phase - v_neutral
    v_c = v_c_phase - v_neutral
    return v_a, v_b, v_c
```

```python
# time series
t = np.linspace(0, 10, 1000)
theta = t

def FOC_loop(t, theta):
    # phase data
    i_a = np.cos(t)
    i_b = np.cos(t + 2./3. * 2 * np.pi)
    i_c = np.cos(t - 2./3. * 2 * np.pi)

    i_alpha, i_beta = FOC_clarkTransform(i_a, i_b, i_c)
    i_q, i_d = FOC_parkTransform(i_alpha, i_beta, np.sin(theta), np.cos(theta))

    v_q = 0 - i_q
    v_d = 0 - i_d

    v_alpha, v_beta = FOC_invParkTransform(v_q, v_d, np.sin(theta), np.cos(theta))

    v_a, v_b, v_c = FOC_invClarkSVPWM(v_alpha, v_beta)

    return i_a, i_b, i_c, i_alpha, i_beta, i_q, i_d, v_q, v_d, v_alpha, v_beta, v_a, v_b, v_c

i_a, i_b, i_c, i_alpha, i_beta, i_q, i_d, v_q, v_d, v_alpha, v_beta, v_a, v_b, v_c = FOC_loop(t, theta)



fig = plt.figure()
plt.subplot(3, 2, 1)
i_a_line, = plt.plot(t, i_a, label="I_a")
i_b_line, = plt.plot(t, i_b, label="I_b")
i_c_line, = plt.plot(t, i_c, label="I_c")
plt.legend(loc="upper right")
plt.title("Phase Current")

plt.subplot(3, 2, 3)
i_alpha_line, = plt.plot(t, i_alpha, label="I_")
i_beta_line, = plt.plot(t, i_beta, label="I_")
plt.legend(loc="upper right")
plt.title("Orthogonalized Current (Clarke Transform)")

plt.subplot(3, 2, 5)
i_q_line, = plt.plot(t, i_q, label="I_q")
i_d_line, = plt.plot(t, i_d, label="I_d")
plt.legend(loc="upper right")
plt.title("Linearized Current (Parke Transform)")

plt.subplot(3, 2, (2, 4))
i_quiver = plt.quiver(0, 0, np.cos(theta[0]), np.sin(theta[0]), scale=3)

plt.tight_layout()
plt.show()

n_frames = 100

def drawframe(n):
    base_t = np.linspace(0, 10, 1000)
    dt = ((2 * np.pi) * (n / n_frames))
    theta = base_t + dt
    i_a, i_b, i_c, i_alpha, i_beta, i_q, i_d, v_q, v_d, v_alpha, v_beta, v_a, v_b, v_c = FOC_loop(base_t + dt, theta)
    i_a_line.set_data(base_t, np.flip(i_a))
    i_b_line.set_data(base_t, np.flip(i_b))
    i_c_line.set_data(base_t, np.flip(i_c))
    i_alpha_line.set_data(base_t, np.flip(i_alpha))
    i_beta_line.set_data(base_t, np.flip(i_beta))
    i_q_line.set_data(base_t, np.flip(i_q))
    i_d_line.set_data(base_t, np.flip(i_d))
    i_quiver.set_UVC(np.cos(theta[0]), np.sin(theta[0]))
    #pt1.set_data(y1[0],y2[0])
    #txt_title.set_text('Frame = {0:4d}'.format(n))
    return (i_a_line, i_b_line, i_c_line, i_quiver)

# blit=True re-draws only the parts that have changed.
anim = animation.FuncAnimation(fig, drawframe, frames=100, interval=20, blit=True)
```

# ADPLL

```python
import math
import random

import matplotlib.pyplot as plt
```

```python
f_sample = 8000 / 4

# generate reference signal
t = np.arange(0, 1, 1 / f_sample)

ref_readings = []
noisy_readings = []
angle = 0
for i in t:
    ref_readings.append(angle)
    noisy_readings.append(angle + 6 * random.random() - 3)
    if i < .5:
        angle = 10 * np.sin(i * 40)
    elif i < .88:
        angle += 0.1 * i
    elif i < .9:
        angle += -2 * i
    else:
        angle += 0
```

```python
filter_bandwidth = 400 / 4
w3db = (1. / f_sample) * 2 * np.pi * filter_bandwidth
filter_k_p = .5 * (2 * w3db)
filter_k_i = .5 * (w3db * w3db)

def ADPLL(position_measured, position_prev, filter_integral):
    position_error = position_measured - position_prev
    filter_integral += filter_k_i * position_error
    position_tune = filter_k_p * position_error + filter_integral

    position = position_prev + position_tune
    return position, filter_integral


adpll_readings = []

filter_integral = 0
position = 0
for position_measured in noisy_readings:
    position, filter_integral = ADPLL(position_measured, position, filter_integral)
    adpll_readings.append(position)
```

```python
iir_filter_alpha = 1 - math.e**(-2 * math.pi * (filter_bandwidth/f_sample))
print(iir_filter_alpha)
iir_readings = []
position = 0
for position_measured in noisy_readings:
    position += iir_filter_alpha * (position_measured - position)
    iir_readings.append(position)
```

```python
filter_bandwidth = 400 / 4
w3db = (1. / f_sample) * 2 * np.pi * filter_bandwidth
filter_k_p = .5 * (2 * w3db)
filter_k_i = .5 * (w3db * w3db)


pidiir_readings = []
position = 0
position_int = 0
for position_measured in noisy_readings:
    error = (position_measured - position)
    position_int += filter_k_i * error
    position = filter_k_p * error + position_int
    pidiir_readings.append(position)
```

```python

plt.plot(t, noisy_readings, label="ref")
plt.plot(t, iir_readings, label="1stOrderIIR")
plt.plot(t, adpll_readings, label="ADPLL")
plt.plot(t, pidiir_readings, label="pidiir")
plt.title("cutoff = 400Hz / 8000Hz")
plt.legend()
plt.show()
```

```python
cpr = 1<<14

N = 1000

# frequency of ref signal (Hz)
f_ref = 10.

# sampling rate (Hz)
f_sample = 8000.
dt = 1 / f_sample

t = [n * dt for n in range(N)]
ref_sig = [10000*(2 * f_ref * n * dt) + 20 * random.random() for n in range(N//2)] + [-1000 + 10000*(5 * f_ref * n * dt) + 20 * random.random() for n in range(N//2)]
ref_sig = [s % cpr for s in ref_sig]

k_nco = 1 / 4096.

w3db = dt * 2 * math.pi * 40

k_p = 2 * w3db / (2 * k_nco)
k_i = w3db ** 2 / (2 * k_nco)

print(k_p, k_i)

prev_u = 0
vtune_int = 0

u_arr = []
vtune_arr = []
pe_arr = []


# floating-point filter
for n in range(N):
    # Phase Detector
    ref_phase = ref_sig[n]
    phase_error = ref_phase - prev_u
    phase_error = 2 * ((phase_error + cpr / 2.) % cpr - cpr / 2.)

    # Loop Filter
    vtune_int += k_i * phase_error
    vtune = phase_error * k_p + vtune_int

    # NCO
    u = prev_u + vtune * k_nco
    u = u % cpr
    prev_u = u

    u_arr.append(u)
    vtune_arr.append(vtune)
    pe_arr.append(phase_error)
'''
# int filter
for n in range(N):
    # Phase Detector
    ref_phase = int(ref_sig[n])
    phase_error = int(ref_phase - prev_u)
    phase_error = int(2 * ((phase_error + cpr // 2) % cpr - cpr // 2))

    # Loop Filter
    vtune_int += int(k_i * phase_error)
    vtune = int(phase_error * k_p + vtune_int)

    # NCO
    u = int(prev_u + vtune * k_nco)
    u = int(u % cpr)
    prev_u = u

    u_arr.append(u)
    vtune_arr.append(vtune)
    pe_arr.append(phase_error)
'''

# IIR filter
signal = 0
iir_signal = []
for n in range(N):
    iir_signal.append(signal)
    signal = 0.9 * signal + 0.1 * ref_sig[n]
    

plt.plot(t, pe_arr, label="phase_error")
plt.legend()
plt.show()

plt.plot(t, ref_sig, label="ref")
plt.plot(t, u_arr, label="u")
plt.plot(t, iir_signal, label="iir")
plt.legend()
plt.show()
```
